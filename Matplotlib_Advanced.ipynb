```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('/content/iris.csv')
iris['Species'] = iris['Species'].replace({'Iris-setosa': 0 , 'Iris-virginica': 1 , 'Iris-versicolor': 2})
iris
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c = iris['Species'],cmap= 'jet')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar()
```

3D Line Plot

```python
x = [10,20,50]
y = [90,40,30]
z = [80,40,60]

fig = plt.figure()

ax = plt.subplot(projection = '3d')

ax.scatter3D(x,y,z,s=[100,100,100])
ax.plot3D(x,y,z,color='red')
```

3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx , yy = np.meshgrid(x,y)
z = xx**2 + yy**2
```

```python
fig = plt.figure(figsize=(15,10))

ax = plt.subplot(projection = '3d')
m = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(m)
```

```python
z= np.sin(xx)**2 + np.cos(xx)**2
fig = plt.figure(figsize=(15,10))

ax = plt.subplot(projection = '3d')
m = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(m)
```

Contour Plots

```python
z = xx**2 + yy**2

fig = plt.figure()

ax = plt.subplot()
m = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(m)
```

```python
z = xx**2 + yy**2

fig = plt.figure()

ax = plt.subplot()
m = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(m)
```
