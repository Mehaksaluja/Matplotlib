```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line Graph
price = [20000,35000,30000,68000,60000,70000]
year = [2015,2016,2017,2018,2019,2020]

plt.plot(year,price)
```

```python
batsman = pd.read_csv('/content/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Giving colors
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# Changing styleof Line
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed',linewidth='3')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='.')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
# markersize change the size of mark
```

```python
# Label
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='.',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D',label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
```

```python
# If their are outliers present in graph than the other elements may show like a straight line
price = [20000,35000,30000,68000,60000,70000,10000000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
# so to avoid it we can make some changes according to us
plt.ylim(0,100000)
plt.xlim(2016,2019)
```

```python
# If data is not visibble properly we can add grids in graph
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='.',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D',label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid()
```

```python
# The same code we written above may not run in code editors until we add show function in it
plt.plot(batsman['index'],batsman['V Kohli'],color='green',marker='.',label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',marker='D',label='Rohit')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend()
plt.show()
```

Bar Chart

```python
childern = [10,20,30,40,10,30]
colors = ['red','blue','green','yellow','pink','purple']

plt.bar(colors,childern)
```

```python
# Horizontal Bar Chart

childern = [10,20,30,40,10,30]
colors = ['red','blue','green','yellow','pink','purple']

plt.barh(colors,childern)
```

```python
df = pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
childern = [10,20,30,40,10,30]
colors = ['red red red red red red','blue blue blue blue blue blue','green','yellow','pink','purple']
# To ignore overlapping of words

plt.bar(colors,childern)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plt.bar(df['batsman'],df['2015'],color = 'blue',label='2015')
plt.bar(df['batsman'],df['2016'],bottom = df['2015'],color='pink',label='2016')
plt.bar(df['batsman'],df['2017'],bottom = (df['2015'] + df['2016']),color = 'purple',label='2017')

plt.legend()
plt.show()
```
